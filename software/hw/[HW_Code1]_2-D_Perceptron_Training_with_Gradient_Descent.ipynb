```python
### Training with manually updating W with "Backward" ###

import torch
#from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim




data = [(1.0,2.1,3.0), (2.0, 3.5, 6.0), (3.0, 3.0, 9.0), (4.0, 2.1, 12.0), (5.0, 7.2, 15.0), (6.0, 10.1, 18.0)]

class Net(nn.Module):
    def __init__(self):  # define Neural Net structure
        super(Net, self).__init__()  # initialise with inherit from super class
        self.fc1 = nn.Linear(2,1,bias=False)  # in dim = 2 in our case, out dim, fully connected linear layer with no bias
    def forward(self, x):  # define forward propagation through the network
        x = self.fc1(x)  # input moves through fully connected layer
        return x
        
net = Net()  # instantiate a neural net object as defined earlier

print(net)
print(list(net.parameters()))

#input = torch.randn(1)
#out = net(input)

#def criterion(out, label):
#    return (label - out)**2
criterion = nn.MSELoss()  # loss criterion defined as mean squared error


optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
#optimizer = optim.Adam(net.parameters(), lr=0.005)

verbose = 0
num_epochs = 1000
for epoch in range(num_epochs): 
    for i, current_data in enumerate(data):
        #print(f"i: {i}, Data: {current_data}")
        X_in_1, X_in_2, Y_target = current_data
        #print(f"X_1 : {X_in_1}, X_2 : {X_in_2}, Y : {Y_target}")
        X_in, Y_target = torch.FloatTensor([X_in_1, X_in_2]), torch.FloatTensor([Y_target])
        #print(f"X input is {X_in}, Y output is {Y_target}")
        optimizer.zero_grad()
        net_outputs = net(X_in)
        loss = criterion(net_outputs, Y_target)
        loss.backward()
        optimizer.step()
    if verbose == 0:
        if epoch <= 10 or epoch >= (num_epochs - 10):
            print("Epoch {} - loss: {}".format(epoch, loss))
    else:
        print("Epoch {} - loss: {}".format(epoch, loss))

for i, current_data in enumerate(data):
    X_in_1, X_in_2, Y_target = current_data
    X_in, Y_target = torch.FloatTensor([X_in_1, X_in_2]), torch.FloatTensor([Y_target])
    output = net(torch.FloatTensor(X_in))
    print("When x = {}, y = {}".format(X_in, output))
```
